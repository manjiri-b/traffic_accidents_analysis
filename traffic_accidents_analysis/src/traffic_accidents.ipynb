4. initial column (or row) clean-up

```python
#remove at least two columns

#Evidence of why Location column should be removed.
print("Let's observe the first row:")
print(df['LATITUDE'][0], df['LONGITUDE'][0], df['LOCATION'][0])

#Removing this coz same information in longitude and latitude column
del df['LOCATION']

#SHOW CODE TO SHOW WHY U DELETE

#The column numbers of person injured and numbers of person killed incorporates number of pedestrians, cyclists, motorist
del df['NUMBER OF PEDESTRIANS INJURED']
del df['NUMBER OF PEDESTRIANS KILLED']
del df['NUMBER OF CYCLIST INJURED']
del df['NUMBER OF CYCLIST KILLED']
del df['NUMBER OF MOTORIST INJURED']
del df['NUMBER OF MOTORIST KILLED']

#SHOW CODE TO SHOW WHY U DELETE

#Transforming columns (NaN Values replaced by 0)
df["ZIP CODE"] = df["ZIP CODE"].fillna(0).apply(np.int64)

#RENAMING COLUMNS
df.rename(columns={"CRASH DATE": "crash_date", "CRASH TIME": "crash_time", "ZIP CODE": "zip_code", 
                   "ON STREET NAME": "street", "NUMBER OF PERSONS INJURED": "people_injured", 
                   "NUMBER OF PERSONS KILLED": "people_killed"}, inplace=True)

#reason: makes columns easier to refer to and work with as decreases chances of syntax error, for example by missing 
#the space between the words, when referring to these columns 
```

Columns deleted:

2. Columns 12-17 (index) were also deleted as the information they provide is already incorporated in the information provided by the number of persons injured and number of persons killed columns. These deleted columns just add more detail which is not necessary for our analysis.

Renaming columns: "CRASH DATE", "CRASH TIME", "ZIP CODE","ON STREET NAME", "NUMBER OF PERSONS INJURED", "NUMBER OF PERSONS KILLED" were renamed.

8. What are the distributions of accidents based on the geo location (latitude & longitude)?

```python
#Subseting the data by borough (for magnitude) using loc (dropping the NaN values using .dropna())
#Removing all values where Latitude and Longitude = 0 (SUGGESTS the authors of the data have replaced)
subset_manhattan_ll = df.loc[((df['BOROUGH'] == 'MANHATTAN') & (df['LATITUDE'] != 0) & (df['LONGITUDE'] != 0)), ['LATITUDE','LONGITUDE']].dropna()
subset_brooklyn_ll = df.loc[((df['BOROUGH'] == 'BROOKLYN') & (df['LATITUDE'] != 0) & (df['LONGITUDE'] != 0)), ['LATITUDE','LONGITUDE']].dropna()
subset_queens_ll = df.loc[((df['BOROUGH'] == 'QUEENS') & (df['LATITUDE'] != 0) & (df['LONGITUDE'] != 0)), ['LATITUDE','LONGITUDE']].dropna()
subset_bronx_ll = df.loc[((df['BOROUGH'] == 'BRONX') & (df['LATITUDE'] != 0) & (df['LONGITUDE'] != 0)), ['LATITUDE','LONGITUDE']].dropna()
subset_statenisland_ll = df.loc[((df['BOROUGH'] == 'STATEN ISLAND') & (df['LATITUDE'] != 0) & (df['LONGITUDE'] != 0)), ['LATITUDE','LONGITUDE']].dropna()

#Plotting the scatter plots
plt.scatter(subset_manhattan_ll['LATITUDE'],subset_manhattan_ll['LONGITUDE'], c = 'red', label = "Manhattan")
plt.scatter(subset_brooklyn_ll['LATITUDE'],subset_brooklyn_ll['LONGITUDE'],  c = 'blue', label = "Brooklyn")
plt.scatter(subset_queens_ll['LATITUDE'],subset_queens_ll['LONGITUDE'], c = 'green', label = "Queens")
plt.scatter(subset_bronx_ll['LATITUDE'],subset_bronx_ll['LONGITUDE'], c = 'coral', label = "Bronx")
plt.scatter(subset_statenisland_ll['LATITUDE'],subset_statenisland_ll['LONGITUDE'], c = 'purple', label = "Staten Island")
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Geo Location Distribution of Accidents')
plt.legend()
plt.show()
```

The data for each borough was subsetted using .loc[] function. 3 conditions were specified in the function. 1. Only observations of the respective borough should be kept through df['BOROUGH'] == respective borough name. 2. The data in the latitude column cannot be 0 and 3. the data in the longitude column cannot be 0. These latter two conditions were specified for cleaning up the data (aforementioned reason in question 3. in other cleaning necessary section). After this the .scatter() function with latitude column as x and longitude column as y was used to plot the data points for each borough. Keyword 'c' was used to specify the colors of the data points and 'label' was used to specify the name of the borough being plotted, which was then shown through the plt.legend() function. X-axis title, Y-axis title and main title were specified using plt.xlabel(), plt.ylabel() and plt.title() respectively.

## 9. shows the covariance between each pair of the columns ##

```python
#Columns I think we should calculate the covariance for
print(df['zip_code'].cov(df['people_injured']))
print(df['zip_code'].cov(df['people_killed']))
print(df['people_killed'].cov(df['people_injured']))
```

Columns chosen: 
- Zipcode and People Injured: wanted to see if there is a relationship between zipcode and number of people injured. Can be useful in answering if in certain areas people tend to have more accidents, hence more injuries.
- Zipcode and People Killed:  wanted to see if there is a relationship between zipcode and number of people killed for the same reason as the previous bullet point
- People Injured and People Killed: wanted to explore the relationship between people injured and people killed. They havee a covariance of 0.0006172 which is very near 0. Hence, it indicates there is no linear relationship between the two varibles. This makes sense as two events are disjoint. One can either be injured or killed.

Documentation:

10. Which month did the most number of accidents occur?

```python
#Format: mm/dd/yy 
crash_dates =  [int(x.split('/')[0]) for x in df['crash_date']]
crash_dates_series = pd.Series(crash_dates).value_counts(sort=False)

crash_dates_series.plot.bar(rot=0)
plt.xlabel("Month")
plt.ylabel("Number of Accidents")
plt.title("Accidents By Month")
plt.show()
```

Janurary (month 1) had the most number of accidents. The number of accidents decreased from Januray onwards until it hit a lull at the month of April (lowest number of accidents). After April the number of accidents started increasing again.

Clean up data process:
List comprehension was used to collect the months from the crash date column. A for loop was used to iterate over each row in the column (iterator variable name: 'x'). Since the date is in "mm/dd/yy" format, .split('/') was used to break the string down into a list ["mm","dd","yy"]. The month is now the first element (index 0) of the list so x.split('/')[0] was used to isolate it. Then since the month is a string to convert it into int, int was used around the entire expression.
